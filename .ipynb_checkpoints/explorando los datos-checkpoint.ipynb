```python
X_train = pd.read_csv("./data/X_train_espanhol.csv", index_col=0)
print(X_train.shape)
X_train.head()
```

```python
def processing(doc):
    return doc.lower()
```

```python
import re, unicodedata, inflect
import spacy
nlp = spacy.load('es_core_news_sm')

def token(doc):
    words = nlp(doc)
    r = []
    for word in words:
        new_word = unicodedata.normalize('NFKD', word.lemma_).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        new_word = re.sub(r'[^\w\s]', '', new_word)
        new_word = re.sub(r'\d','',new_word)
        if len(new_word) > 1:
          #new_word = SnowballStemmer("spanish").stem(new_word)
          r.append(new_word)
    return r
```

```python
token("me comí una salchipapa y que baina mas sabrosa, pero no me alcanzó pa la gaseosa")
```

```python
text = X_train.title.values

vectorizer = CountVectorizer(stop_words=nltk.corpus.stopwords.words("spanish"), strip_accents="unicode",
                             preprocessor=processing, decode_error="ignore", analyzer="word", max_features=1550, tokenizer=token)

X_train_esp = vectorizer.fit_transform(text)
print("número de features: ", len(vectorizer.get_feature_names()))
encoder = OrdinalEncoder()
y_train_esp = X_train.category.values.reshape(-1, 1)
encoder.fit(y_train_esp)
y_esp_enc = encoder.transform(y_train_esp)
print(X_train_esp.shape, y_esp_enc.shape)
```

```python
np.save("./data/matrix_espanhol", X_train_esp.toarray())
```

```python
model = Sequential()
model.add(Dense(300, input_shape=(X_train_esp.shape[1],), name="dense1", activation="sigmoid"))
model.add(Dropout(.1))
model.add(Dense(600,  name="dense2", activation="tanh"))
model.add(Dense(len(np.unique(y_esp_enc)), activation="softmax"))
```
